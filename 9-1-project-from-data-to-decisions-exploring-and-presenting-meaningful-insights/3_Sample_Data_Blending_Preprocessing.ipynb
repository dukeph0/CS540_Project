### Data File 1 dataframe
Here we will load the data_file_1 example data, then pivot and clean it in preparation for merging with the other two selected sample files.

For your submission you will adjust the code to reflect your selected dataset.


```python
data1 = pd.read_csv(data_file_1)
data1.head()
```

```python
data1_wide = data1.pivot(index=['FIPStxt','State','Area_Name'], columns='Attribute', values='Value')
data1_wide.info()
```

```python
data1_wide.dropna(thresh=40, axis=1, inplace=True)
data1_wide.info()
```

```python
data1_wide.fillna(value=0, inplace=True)
data1_wide
```

### Data File 2 dataframe
Here we will load the data_file_2 example data, then pivot and clean it in preparation for merging with the other two selected sample files.

For your submission you will adjust the code to reflect your selected dataset.

```python
data2 = pd.read_csv(data_file_2)
data2.head()
```

```python
data2.rename(columns={'Area name': 'Area_Name', 'FIPS Code': 'FIPStxt'}, inplace=True)
data2.head()
```

```python
data2_wide = data2.pivot(index=['FIPStxt','State','Area_Name'], columns='Attribute', values='Value')
data2_wide.info()
```

```python
data2_wide
```

### Data File 3 dataframe
If you choose not to included a third file for merging, remove or comment out the blocks for data3.

Here we will load the data_file_3 example data, then pivot and clean it in preparation for merging with the other two selected sample files.

For your submission you will adjust the code to reflect your selected dataset.

```python
data3 = pd.read_csv(data_file_3)
data3.head()
```

```python
data3.rename(columns={'FIPS_Code': 'FIPStxt'}, inplace=True)
data3.head()
```

```python
data3_wide = data3.pivot(index=['FIPStxt','State','Area_Name'], columns='Attribute', values='Value')
data3_wide.info()
```

```python
data3_wide.dropna(thresh=40, axis=1, inplace=True)
data3_wide.info()
```

```python
data3_wide.fillna(value=0, inplace=True)
data3_wide
```

### Data Merging
Here is the example of a merge of the three sample files. You will need to adjust these blocks for how you wish to merge your selected data files.

<div class="alert alert-block alert-danger">
    <span style="color:black"><b><i>Example Data:</i></b> Keep in mind that the output below are from the sample data and will need replaced before running the block with your dataset for how you need merge your dataset.</span>

```python
merged_data_1_2 = pd.merge(data1_wide, data2_wide, on=['FIPStxt', 'State', 'Area_Name'])
merged_data_1_2
```

```python
final_merged_data = pd.merge(data3_wide, merged_data_1_2, on=['FIPStxt', 'State', 'Area_Name'])
final_merged_data
```
